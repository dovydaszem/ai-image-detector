```python
import os
import torch
import shutil
from tqdm import tqdm
from IPython.display import FileLink
from diffusers import AutoPipelineForText2Image
```

```python
# Load the pipeline in float16
model = "stabilityai/sdxl-turbo"
pipe = AutoPipelineForText2Image.from_pretrained(model, torch_dtype=torch.float16, variant="fp16")
pipe.enable_model_cpu_offload()
```

```python
start_at = 1000; end_at = 2000
captions_fname = "/kaggle/input/captions/captions.txt"
!rm -rf /kaggle/working/gen
os.makedirs("/kaggle/working/gen", exist_ok=True)

with open(captions_fname, "r") as f:
    for i, line in enumerate(f):
        if i < start_at:
            continue
        if i >= end_at:
            break
        prompt = line.rstrip("\n")
        img = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]
        img.save(f"/kaggle/working/gen/{str(i).zfill(4)}.jpg")
```

```python
!rm /kaggle/working/gen.zip
shutil.make_archive("/kaggle/working/gen", 'zip', "/kaggle/working/gen")
FileLink(r'gen.zip')
```
